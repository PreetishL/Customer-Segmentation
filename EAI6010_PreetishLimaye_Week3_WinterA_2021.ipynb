```
import pandas as pd
def get_relative_freq_df():
  text = nltk.corpus.gutenberg.fileids()
  fnames = [str(f) for f in gutenberg.fileids()]
  #Create count tuple
  freq_list = []
  test_list = []
  for fields in gutenberg.fileids():
    fld_txt = nltk.FreqDist(w.lower() for w in gutenberg.words(fields))   
    modals = ['can','could','may','might','must','will','would','should']
    for m in modals:
      freq_list.append(fld_txt[m])
    test_list.append(tuple(freq_list))
    freq_list.clear()
  #Create dataframe
  df = pd.DataFrame(test_list,columns=['can','could','may','might','must','will','would','should'])
  df['total'] = df.sum(axis = 1)
  relative_df = df[modals].div(df.total, axis = 0)
  final_df = relative_df.round(2)
  return final_df


```
